```python

import json

# Specify the path to your JSON file
file_path ='output_file.json'

# Open the JSON file
with open(file_path) as json_file:
    # Parse the JSON data
    data = json.load(json_file)
```

```python
def get_formatted_transcript(paragraphs, get_timestamp=False, get_speaker=False):
    transcript = ''
    for paragraph in paragraphs:
        minutes = paragraph['start'] // 60
        seconds = paragraph['start'] % 60
        time_string = "{:02d}:{:02d}".format(minutes, seconds)
        text = ""
        if get_speaker:
            text += f"<Speaker {paragraph['speaker']}> "
        if get_timestamp:
            text += f"{time_string} " 
        text += f"{paragraph['text']} \n"
        transcript += text
    return transcript

def get_transcribed_text(response):
    results = {'paragraphs': [], 'transcript': '', 'transcript_with_ts': '', 'transcript_with_ts_with_speaker':''}
    paragraphs = response['results']['channels'][0]['alternatives'][0]['paragraphs']['paragraphs']
    for paragraph in paragraphs:
        paragraph_data = {
            'speaker': paragraph['speaker'],
            'start': int(paragraph['start']),
            'end': int(paragraph['end']),
            'text': ' '.join(sentence['text'] for sentence in paragraph['sentences'])
        }
        results['paragraphs'].append(paragraph_data)
    processed_paragraphs = results['paragraphs']
    results['transcript'] = get_formatted_transcript(processed_paragraphs, get_timestamp=False, get_speaker=False)
    results['transcript_with_ts'] = get_formatted_transcript(processed_paragraphs, get_timestamp=True, get_speaker=False)
    results['transcript_with_ts_and_speaker'] = get_formatted_transcript(processed_paragraphs, get_timestamp=True, get_speaker=True)
    return results
```
